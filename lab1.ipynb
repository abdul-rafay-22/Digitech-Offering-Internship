```python
df['total_score'] = df['math score'] + df['reading score'] + df['writing score']
df['average_score'] = df['total_score'] / 3
def get_grade(score):
    if score >= 80: return 'A'
    elif score >= 70: return 'B'
    elif score >= 60: return 'C'
    elif score >= 50: return 'D'
    else: return 'F'
df['performance_grade'] = df['average_score'].apply(get_grade)
df['high_achiever'] = df['average_score'] >= 80
df['lowest_subject'] = df[score_cols].idxmin(axis=1).str.replace(' score', '')
display(df.head(10))
print("Performance Grade Distribution:")
print(df['performance_grade'].value_counts())
print("\nLowest Subject Distribution:")
print(df['lowest_subject'].value_counts())
```

#### Step 14

```python
def score_summary(df, col):
    """
    Returns a dictionary of summary statistics for a given numerical column.
    Includes mean, median, std, min, max, and percentage of values above 70.
    """
    total_count = len(df)
    above_70 = len(df[df[col] > 70]) 
    return {
        'mean': round(df[col].mean(), 2),
        'median': round(df[col].median(), 2),
        'std': round(df[col].std(), 2),
        'min': df[col].min(),
        'max': df[col].max(),
        'pct_above_70': round((above_70 / total_count) * 100, 2)
    }
def plot_distribution(df, col, color):
    """
    Plots a histogram with a KDE curve and a vertical line for the mean.
    """
    plt.figure(figsize=(8, 5))
    sns.histplot(df[col], kde=True, color=color, bins=20)
    plt.axvline(df[col].mean(), color='red', linestyle='dashed', linewidth=2, label=f'Mean: {df[col].mean():.2f}')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()

def compare_groups(df, group_col, score_col):
    """
    Returns a sorted DataFrame showing mean scores and % above 70 by group.
    """
    grouped = df.groupby(group_col).apply(
        lambda x: pd.Series({
            'mean_score': round(x[score_col].mean(), 2),
            'pct_above_70': round((len(x[x[score_col] > 70]) / len(x)) * 100, 2)
        })
    ).sort_values(by='mean_score', ascending=False)
    return grouped
print("Score Summary (Math):")
print(score_summary(df, 'math score'))
plot_distribution(df, 'reading score', 'purple')
print("Group Comparison (Lunch vs Writing Score):")
display(compare_groups(df, 'lunch', 'writing score'))
```

#### Step 15

```python
math_arr = df['math score'].values
read_arr = df['reading score'].values
write_arr = df['writing score'].values
arrays = {'Math': math_arr, 'Reading': read_arr, 'Writing': write_arr}
print("NumPy Manual Computations")
for name, arr in arrays.items():
    print(f"\n--- {name} ---")
    mean = np.mean(arr)
    std = np.std(arr)
    print(f"Mean: {mean:.2f}, Median: {np.median(arr)}, Std: {std:.2f}")
    print(f"Var: {np.var(arr):.2f}, Min: {np.min(arr)}, Max: {np.max(arr)}")
    z_scores = (arr - mean) / std
    outliers = np.sum((z_scores > 2) | (z_scores < -2))
    print(f"Students with |z| > 2: {outliers}")
    
    # Pass Rate
    pass_rate = np.mean(arr > 70) * 100
    print(f"Pass Rate (>70): {pass_rate:.2f}%")
percentiles = [10, 25, 50, 75, 90]
perc_dict = {name: np.percentile(arr, percentiles) for name, arr in arrays.items()}
perc_df = pd.DataFrame(perc_dict, index=[f'{p}th' for p in percentiles])
print("\nPercentiles:")
display(perc_df)
top_10_math_idx = np.argsort(math_arr)[-10:][::-1]
print(f"\nIndices of Top 10 Math Scorers: {top_10_math_idx}")
```
